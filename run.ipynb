```python
from datasets import load_dataset

ds = load_dataset("tmnam20/Vietnamese-News", "dantri")
```

```python
len(ds["train"]["description"])
```

```python
# prompt: sort the description contain number like "1" , "5.000" , "6,6"

import re

def has_number(text):
  return bool(re.search(r'\d', text))

# Filter descriptions containing numbers
descriptions_with_numbers = [desc for desc in ds["train"]["description"] if has_number(desc)]

# Sort the filtered descriptions
sorted_descriptions = sorted(descriptions_with_numbers)

# Print the sorted descriptions

```

```python
messages = [
    {"role": "user", "content": "Who are you?"},
]
```

```python
import pandas as pd
```

```python
sorted_df = pd.DataFrame(sorted_descriptions)
sorted_df.columns = ["original"]
```

```python
sorted_df.head()
```

```python
def give_prompt(p):
    prompt = f"""Chuyển đổi các con số từ dạng chữ viết sang dạng chữ nói bằng tiếng Việt. Các con số đưa ra dưới đây có thể là số đếm, giá trị tiền tệ hoặc ngày tháng. Làm theo các hướng dẫn này:

        Ví dụ Chuyển đổi số đếm:
        - 100 thành “một trăm”
        - 15.000 thành “mười năm nghìn”
        Ví dụ Chuyển đổi số tiền tệ:

        - 13 tỷ USD thành “mười ba tỷ đô la”
        - 500.000 VNĐ thành “năm trăm nghìn đồng”
        Ví dụ chuyển đổi ngày tháng:

        - Ngày 15/8 thành “ngày mười năm tháng tám”
        - Ngày 1/1 thành “ngày một tháng một”
        Dữ liệu đầu vào sẽ được đưa ra trong đoạn văn sau:
        {p}
        Dữ liệu đầu ra cần được làm một cách hoàn chỉnh, không gồm các kí tự đặc biệt, và chỉ chứa nội dung của câu gốc.
        Hãy chuyển đổi các con số trong đoạn văn trên thành dạng chữ nói bằng tiếng Việt theo các hướng dẫn trên."""
    return prompt
```

```python
test_df   = sorted_df["original"].head(10)
```

```python
test_list = test_df.tolist()
```

```python
for p in test_list:
    print(p)
```

```python
url = 'http://localhost:11434/api/chat'
response_list = []
for p in test_list:

    payload = {
        "model": "gemma2:2b",
        "temperature": 0.6,
        "stream": False,
        "messages": [
            {"role": "system", "content": "You are an AI assistant!"},
            {"role": "user", "content": give_prompt(p)}
        ]
    }

    response = requests.post(url, json=payload)
    message_str = response.content.decode('utf-8')
    message_dict = json.loads(message_str)
    response_list.append(message_dict['message']['content'])

```

```python
result = dict(zip(test_list, response_list))
result_df = pd.DataFrame(result.items(), columns=['original', 'response'])
```
